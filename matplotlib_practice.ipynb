```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
x = np.linspace(0, 5, 40)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black')
plt.axis('equal')
plt.title("Simple plot");
```
